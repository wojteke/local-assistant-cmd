```python
import matplotlib.pyplot as plt
import numpy as np

models = ["GPT2", "GPT2 Lora", "GPT2 Medium",
          "GPT2 Medium Lora", "GPT2 Large", "GPT2 Large Lora"]

# Set the size of the figure
plt.figure(figsize=(10, 6))

# Setting the positions and width for the bars
bar_width = 0.35
positions_a = np.arange(len(models))
positions_b = [x + bar_width for x in positions_a]

color1 = 'teal'
color2 = 'coral'

# Plotting the bars
bars1 = plt.bar(positions_a, bleu2_100_max_length,
                width=bar_width, color=color1, label='Max Length')
bars2 = plt.bar(positions_b, bleu2_100_max_new_tokens,
                width=bar_width, color=color2, label='Max New Tokens')

# Adding the x-axis labels
plt.xlabel('Model', fontsize=12)
plt.ylabel('Wynik BLEU', fontsize=12)
plt.title('Wyniki BLEU dla kadego modelu')

# Adding legends
plt.legend(loc='lower right')

# Adding xticks
plt.xticks([pos + bar_width / 2 for pos in positions_a],
              models, rotation=15)

# Adding the values above the bars
for bars in (bars1, bars2):
     for bar in bars:
         height = bar.get_height()
         plt.text(bar.get_x() + bar.get_width() / 2.0, height,
                  f'{height:.2f}', ha='center', va='bottom')
# Show plot

plt.savefig("bleu.png", dpi=300, bbox_inches='tight')

plt.tight_layout()
plt.show()
```
